```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
#multiplots
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig = plt.figure() #a figure object is created =  a blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # these create axes at given positions [x0,y0,width,height]
axes.plot(x,y) # plot the graph
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes2 = fig.add_axes([0,0,1,1])
axes2.plot(y,x)
axes1 = fig.add_axes([0.55,0.1,0.4,0.3])
axes1.plot(x,y)
```

```python
#oo subplots
fig,axes = plt.subplots(nrows=1,ncols=2) #tuple unpacking
# axes.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout() #beautify plot
```
